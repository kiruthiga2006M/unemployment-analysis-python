```python
import pandas as pd

df = pd.read_csv('/kaggle/input/datasets/pantanjali/unemployment-dataset/unemployment analysis.csv')

df.head()
```

```python
import matplotlib.pyplot as plt

df.iloc[:,2:].mean().plot()

plt.title("Global Unemployment Trend")
plt.xlabel("Year")
plt.ylabel("Unemployment Rate")
plt.show()
```

```python
covid = df[['2019','2020','2021']].mean()
print(covid)

covid.plot(kind='bar')
plt.title("Covid-19 Impact on Unemployment")
plt.show()
```

```python
df.iloc[:,2:].mean().plot(marker='o')
plt.title("Yearly Unemployment Pattern")
plt.show()
```

Conclusion:

The unemployment rate increased significantly during 2020 due to Covid-19 pandemic.
Many countries experienced economic slowdown and job losses.
After 2021, unemployment gradually decreased showing economic recovery.
This analysis helps governments plan employment and economic policies.
